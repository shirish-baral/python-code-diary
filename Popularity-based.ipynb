# Popularity-Based Recommender System

This notebook builds a **popularity-based recommender system** for books. It recommends the **top 50 most popular books** based on high average ratings and a significant number of user ratings.

### **Theory**

1. **Minimum Ratings Threshold:**
   - Only books with at least **250 user ratings** are considered to ensure reliability.

2. **Sorting Criterion:**
   - Books are sorted by **average rating** in descending order.

3. **Top-N Selection:**
   - The **top 50 books** are selected from the sorted list.


### **Output**

- A DataFrame with columns:
  - `Book-Title`
  - `Number of Ratings`
  - `Average Rating`
  - `Book-Author`
  - `Image-URL-M`

This simple yet effective system serves as a baseline for recommending books to new users or in the absence of personalized preferences.

## Popularity based Recommender System

```python
# Highest average ratings books which has atleast 250 ratings
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name
```

```python
# Name of books and number of ratings it has got

num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean(numeric_only = True).reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# Extract only books with number of ratings > 250
popular_df['num_ratings']>=250
popular_df
```

```python
# Books sorted in descending order based on their average ratings
popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
# Top 50 books based on avg_ratings

popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
# 50 number of books
popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title').shape
```
