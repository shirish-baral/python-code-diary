```python
import pandas as pd
import random

# Set random seed for reproducibility
random.seed(42)
np.random.seed(42)

# List of common Nepali names
nepali_names = [
    "Ram", "Shyam", "Hari", "Sita", "Gita", "Krishna", "Radha", "Madan", 
    "Puja", "Manish", "Prakash", "Sunita", "Ramesh", "Anita", "Suman", 
    "Dipesh", "Kiran", "Binita", "Sandesh", "Rekha", "Bikash", "Niraj", 
    "Sarita", "Aarati", "Santosh", "Bimal", "Pooja", "Sanjay", "Sabina", "Rupa"
]

# List of common Nepali food items
food_items = [
    "Momo", "Dal Bhat", "Sel Roti", "Thukpa", "Chowmein", "Sekuwa",
]

# Generate sample data
names = ["Shirish Baral"] + [random.choice(nepali_names) for _ in range(49)]
genders = ["Male"] + [random.choice(["Male", "Female"]) for _ in range(49)]
cities = ["Pokhara"] + [random.choice(["Pokhara","Kathmandu", "Lalitpur", 
                                       "Dharan", "Chitwan", "Bhaktapur", "Butwal", "Nepalgunj"]) for _ in range(49)]
frequencies = [random.choice(["Daily", "Weekly", "Monthly"]) for _ in range(50)]
item = ["Momo"] + [random.choice(food_items) for _ in range(49)]
spends = [random.randint(500, 5000) for _ in range(50)]
# Amounts in Nepali Rupees (NPR)

# Create the DataFrame
food = pd.DataFrame({
    "Name": names,
    "Gender": genders,
    "City": cities,
    "Frequency": frequencies,
    "Food Item": item,
    "Spends (NPR)": spends
})

# Optionally, save the dataset to a CSV file
food.to_csv("food_dataset_nepal.csv", index=False)

food
```
