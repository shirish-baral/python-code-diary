```python
# drop_duplicates Functions
```

```python
# Total number of cities that have hosted IPL
data.drop_duplicates(subset=['city']).shape[0]
```

```python
# To find IPL winner of each season
```

```python
#All final matches
final = data.drop_duplicates('season',keep='last') 
# Winner of each season
winners = final[['season','winner']]
# Winner of each season in asceding order
winners.sort_values('season') 
#In single command
# data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```

```python
# number of times team has won
winners.value_counts('winner') 
```
